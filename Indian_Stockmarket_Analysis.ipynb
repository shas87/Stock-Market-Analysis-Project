# Title : Data analysis and visualization of IPO's 

Importing libraries

```python
# for mathematical computation

import pandas as pd
import numpy as np
```

```python
# for visualization

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# loading dataset
df = pd.read_excel('IPO_Data.xlsx')
```

```python
df.head(10)
```

```python
# to get info about data type of columns
df.info()
```

```python
# changing data type of Date column
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Number of records
len(df)
```

```python
df.head(10)
```

```python
df.describe()
```

```python
# To check null values
df.isnull().sum()
```

```python
df[df['Current_gains'].isnull()]
```

```python
# dropping Current_gains column
df = df.drop('Current_gains', axis='columns') 
```

```python
df.head()
```

```python
# correlation(Heatmap)
v1 = sns.heatmap(df.corr(), annot = True, cmap = 'Blues')
plt.rcParams['figure.figsize'] = 12,8
```

Inferences:

1) High Net-worth Individual(HNI) is very positively correlated to Qualified institutional buyer(QIB).

2) Retail individual investor(RII) is very positively correlated to HNI.

3) The issue price is highly positively correlated to Listing open.

4) Negligible correlation exist between Issue_Size(crores) and Listing_Gains(%). 

```python
# Listing_Gains(%) distribution
v = sns.kdeplot(df['Listing_Gains(%)'])
```

Inference: Most of the stocks have listing gain% between 0-6% 

```python
#Extracting Day, Year, Month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df.head()
```

```python
# Number of IPO launches year wise
sns.set_style('dark')
v2 = sns.displot(df['Year'], bins = 12)
plt.show()
```

```python
len(df[df['Year'] == 2010])
```

Inference: Maximum number of IPO's were listed in year 2010, then in 2021.

```python
# IPO's launch month wise
v3 = sns.displot(df['Month'], bins = 12)
```

Inference: Maximum number of IPO's were listed in the month of October.

```python
# top gainers at the time of listing
df1 = df.sort_values('Listing_Gains(%)',ascending = False).head(10)
df1 = df1.reset_index(drop =True)
df1
```

```python
v4 = sns.barplot(data = df1, x='IPO_Name', y='Listing_Gains(%)')
plt.rcParams['figure.figsize'] = (15,6)
```

Inference: Sigachi Ind has the maximum listing gain of 270.4% at the time of listing. 

```python
df[df['IPO_Name'] == 'Sigachi Ind']
```

```python
# top loosers at the time of listing
df2 = df.sort_values('Listing_Gains(%)').head(10)
df2 = df2.reset_index(drop = True)
df2
```

```python
v5 = sns.barplot(data=df2, x='IPO_Name', y='Listing_Gains(%)')
```

```python
df[df['IPO_Name'] == 'VKS Projects']
```

Inference: VKS Projects is the biggest looser at the time of listing.

```python
# top 10 IPO's in terms of issue size
df3 = df.sort_values('Issue_Size(crores)', ascending = False).head(10)
df3 = df3.reset_index(drop=True)
df3
```

```python
v6 = sns.barplot(data = df3, x = 'IPO_Name', y = 'Issue_Size(crores)') 
```

Inference: One 97(Paytm) has the largest Issue_Size.
